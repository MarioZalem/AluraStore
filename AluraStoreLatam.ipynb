### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de cada tienda sumando la columna 'Precio'
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda_4:,.2f}")
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda
ventas_categoria_tienda_1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_tienda_2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_tienda_3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_tienda_4 = tienda4['Categoría del Producto'].value_counts()

print('Tienda 1 - Ventas por categoría:')
print(ventas_categoria_tienda_1)
print('\nTienda 2 - Ventas por categoría:')
print(ventas_categoria_tienda_2)
print('\nTienda 3 - Ventas por categoría:')
print(ventas_categoria_tienda_3)
print('\nTienda 4 - Ventas por categoría:')
print(ventas_categoria_tienda_4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio de los clientes para cada tienda
calif_prom_tienda_1 = tienda['Calificación'].mean()
calif_prom_tienda_2 = tienda2['Calificación'].mean()
calif_prom_tienda_3 = tienda3['Calificación'].mean()
calif_prom_tienda_4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calif_prom_tienda_1:.2f}")
print(f"Calificación promedio Tienda 2: {calif_prom_tienda_2:.2f}")
print(f"Calificación promedio Tienda 3: {calif_prom_tienda_3:.2f}")
print(f"Calificación promedio Tienda 4: {calif_prom_tienda_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Identificar productos más y menos vendidos en cada tienda y visualizarlos
%pip install matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')  # Consistencia visual
def plot_top_bottom(df, nombre_tienda):
    conteo = df['Producto'].value_counts()
    top = conteo.head(5)
    bottom = conteo.tail(5)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    top.plot(kind='bar', ax=axes[0], color='green')
    axes[0].set_title(f'Top 5 más vendidos - {nombre_tienda}')
    axes[0].set_ylabel('Cantidad vendida')
    axes[0].set_xlabel('Producto')
    bottom.plot(kind='bar', ax=axes[1], color='red')
    axes[1].set_title(f'Top 5 menos vendidos - {nombre_tienda}')
    axes[1].set_ylabel('Cantidad vendida')
    axes[1].set_xlabel('Producto')
    plt.tight_layout()
    plt.show()

plot_top_bottom(tienda, 'Tienda 1')
plot_top_bottom(tienda2, 'Tienda 2')
plot_top_bottom(tienda3, 'Tienda 3')
plot_top_bottom(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_prom_tienda_1 = tienda['Costo de envío'].mean()
envio_prom_tienda_2 = tienda2['Costo de envío'].mean()
envio_prom_tienda_3 = tienda3['Costo de envío'].mean()
envio_prom_tienda_4 = tienda4['Costo de envío'].mean()

print(f"Envío promedio Tienda 1: ${envio_prom_tienda_1:.2f}")
print(f"Envío promedio Tienda 2: ${envio_prom_tienda_2:.2f}")
print(f"Envío promedio Tienda 3: ${envio_prom_tienda_3:.2f}")
print(f"Envío promedio Tienda 4: ${envio_prom_tienda_4:.2f}")
```

# 6. Generando gráfico

```python
# Visualización 1: Ingresos totales por tienda (gráfico de barras)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.bar(nombres, ingresos, color=['#4CAF50', '#2196F3', '#FFC107', '#FF5722'])
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()

# Visualización 2: Distribución de categorías más vendidas en Tienda 2 (barras horizontales)
top_categorias_t2 = ventas_categoria_tienda_2.head(7)
plt.figure(figsize=(8,5))
top_categorias_t2.sort_values().plot(kind='barh', color='#2196F3')
plt.title('Top Categorías Más Vendidas - Tienda 2')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()

# Visualización 3: Calificación promedio por tienda (gráfico de líneas)
calificaciones = [calif_prom_tienda_1, calif_prom_tienda_2, calif_prom_tienda_3, calif_prom_tienda_4]
plt.figure(figsize=(8,5))
plt.plot(nombres, calificaciones, marker='o', linestyle='-', color='#4CAF50')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación Promedio')
plt.xlabel('Tienda')
plt.ylim(0,5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
